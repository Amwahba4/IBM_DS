```python
import requests
import pandas as pd
from pandas import json_normalize

# Step 1: Fetch launch data from SpaceX API
url = 'https://api.spacexdata.com/v4/launches'
response = requests.get(url)
launches_data = response.json()

# Convert to DataFrame
df_launches = pd.DataFrame(launches_data)

# Step 2: Explode the cores and payloads lists
df_launches_exploded = df_launches.explode('cores').explode('payloads')

# Step 3: Normalize the cores data
df_cores_normalized = json_normalize(df_launches_exploded['cores'])

# Step 4: Normalize the payloads data
df_payloads_normalized = json_normalize(df_launches_exploded['payloads'])

# Step 5: Combine all DataFrames
df_combined = df_launches_exploded.drop(columns=['cores']).reset_index(drop=True)

# Concatenate the launch data with normalized cores and payloads
df = pd.concat([df_combined, df_cores_normalized.reset_index(drop=True), df_payloads_normalized.reset_index(drop=True)], axis=1)

# Display the final DataFrame structure
print(df.columns)  # List all columns to verify presence of payloads
print(df)   # Display first few records



```

```python
import requests
import pandas as pd

# Fetch rocket details from SpaceX API
url = 'https://api.spacexdata.com/v4/rockets'
response = requests.get(url)
rockets_data = response.json()

# Convert rocket data to a DataFrame
df_rockets = pd.DataFrame(rockets_data)

# Display columns of the rockets DataFrame
print(df_rockets.columns)

# Display key rocket information, such as rocket ID and name
print(df_rockets[['id', 'name', 'type']])

#rename column
df_rockets.rename(columns={'name': 'booster_version','type' : 'booster_type'}, inplace=True)
df_rockets=df_rockets[['id','booster_version','booster_type']]
df_rockets.head()
```

```python
import requests
import pandas as pd

# Step 1: Get all launches
launches_url = "https://api.spacexdata.com/v4/launches/"
response = requests.get(launches_url)

# Check if the request was successful
if response.status_code == 200:
    launches = response.json()

    # Step 2: Collect all payload IDs
    payload_ids = []
    for launch in launches:
        payload_ids.extend(launch['payloads'])  # Collect payload IDs from each launch

    # Step 3: Fetch details for each unique payload ID
    payloads_data = []
    unique_payload_ids = set(payload_ids)  # Use a set to avoid duplicates

    for payload_id in unique_payload_ids:
        payload_url = f"https://api.spacexdata.com/v4/payloads/{payload_id}"
        payload_response = requests.get(payload_url)

        if payload_response.status_code == 200:
            payload_data = payload_response.json()
            payloads_data.append(payload_data)  # Add the payload details to the list
        else:
            print(f"Failed to retrieve payload details for ID {payload_id}. Status code: {payload_response.status_code}")

    # Step 4: Convert the list of payload details into a DataFrame
    df_payloads = pd.DataFrame(payloads_data)

    # Display the DataFrame
    print(df_payloads)
else:
    print(f"Failed to retrieve launches. Status code: {response.status_code}")
```

```python
df_payloads.rename(columns={'name': 'payload_name','mass_kg' : 'payload_mass_kg'}, inplace=True)
df_payloads=df_payloads[['id','payload_name','payload_mass_kg','orbit','customers']]
df_payloads.head()
```

```python
import requests
import pandas as pd

# Step 1: Fetch the launches data to get launchpad IDs
launches_url = "https://api.spacexdata.com/v4/launches"
launch_data = requests.get(launches_url).json()

# Step 2: Extract unique launchpad IDs from launch data
launchpad_ids = set(launch['launchpad'] for launch in launch_data)

# Step 3: Function to get the launchpad details using the ID
def get_launchpad_name(launchpad_id):
    launchpad_url = f"https://api.spacexdata.com/v4/launchpads/{launchpad_id}"
    response = requests.get(launchpad_url).json()
    return {'id': launchpad_id, 'name': response['name'], 'locality': response['locality']}

# Step 4: Fetch the names of launchpads using their IDs
launchpad_details = [get_launchpad_name(launchpad_id) for launchpad_id in launchpad_ids]

# Convert to DataFrame for easier handling
df_launchpads = pd.DataFrame(launchpad_details)
df_launchpads = df_launchpads.rename(columns={'name': 'launch_site'})


df_launchpads=df_launchpads[['id','launch_site','locality']]
df_launchpads.head()
```

#Using SQL quries to manipulate data

```python
import json
# Convert complex types (e.g., lists, dictionaries) to strings
df_final_clean['customers'] = df_final_clean['customers'].apply(lambda x: json.dumps(x) if isinstance(x, (list, dict)) else x)
df_final_clean['customers'] = df_final_clean['customers'].str.replace(r'[\[\]]', '', regex=True)
df_final_clean['customers'] = df_final_clean['customers'].str.replace('"', '', regex=False)
df_final_clean.head()
```

```python
import sqlite3
conn = sqlite3.connect(':memory:')
df_final_clean.to_sql('missions', conn, index=False, if_exists='replace')
cur = conn.cursor()
```

1-display the names of unique launch sites

2-display 5 records where launch site begin with 'cca'

```python
query = "SELECT launch_site FROM missions where launch_site like'CCA%';"
select_cca = pd.read_sql_query(query, conn)
print(select_cca)
```

3-the total payload mass carried by boosters launched by NASA

```python
query = "SELECT sum(payload_mass_kg) FROM missions where customers='NASA (CRS)';"
totalpayload_nasa = pd.read_sql_query(query, conn)
print(totalpayload_nasa)
```
